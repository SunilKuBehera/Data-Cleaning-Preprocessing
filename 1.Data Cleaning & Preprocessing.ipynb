```python
# Set Seaborn style
sns.set(style="whitegrid")

# Define color
color = ['skyblue']

# Get numeric columns
numeric_cols = df.select_dtypes(include='number').columns
num_cols = len(numeric_cols)

# Set up subplot grid
cols_per_row = 3
rows = (num_cols + cols_per_row - 1) // cols_per_row
plt.figure(figsize=(6 * cols_per_row, 5 * rows))

for idx, col in enumerate(numeric_cols, 1):
    plt.subplot(rows, cols_per_row, idx)
    
    # Choose a vibrant color for each boxplot
    colors = color[(idx - 1) % len(color)]
    
    sns.boxplot(data=df, x=col, color=colors)
    plt.title(f"Boxplot of {col}", fontsize=12)
    plt.xlabel(col)
    plt.tight_layout()

plt.suptitle("Boxplots to Detect Outliers", fontsize=16, y=1.02)
plt.show()
```

```python
def find_outliers(series, method="IQR", factor=1.5):
    if method == "IQR":
        Q1 = series.quantile(0.25)
        Q3 = series.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - factor * IQR
        upper_bound = Q3 + factor * IQR
        return (series < lower_bound) | (series > upper_bound)
    else:
        raise ValueError("Unsupported method. Use 'IQR'.")

# Loop through numeric columns and print outlier counts
for col in df.select_dtypes(include="number").columns:
    mask = find_outliers(df[col].dropna())
    count = mask.sum()
    total = len(df[col])
    print(f"{col} : {count} Outliers out of {total} values , Outlier Percentage : ({count/total:.2%})")
```

## 5. Data Preprocessing 

-  Missing Value Treatment

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Fill missing values in categorical columns with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Fill missing values in numerical columns
for col in df.select_dtypes(include='float64').columns:
    if df[col].isnull().sum() > 0:
        # Use median if data is skewed, otherwise mean
        skewness = df[col].skew()
        if abs(skewness) > 1:
            df[col].fillna(df[col].median(), inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()  # Should show all zeros
```

- Outlier Treatment

```python
# Outlier treatment using IQR capping
# Create a copy of the original DataFrame
df_before = df.copy()

# Apply IQR-based outlier capping
for col in df.select_dtypes(include='float64').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].apply(lambda x: 
                            lower_bound if x < lower_bound else 
                            upper_bound if x > upper_bound else x)
```

```python
# Plot before and after boxplots in subplots
cols_to_plot = ['GDP', 'Total expenditure']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 16))
fig.suptitle('Outlier Treatment: Before vs After', fontsize=16)

for i, col in enumerate(cols_to_plot):
    # Before
    sns.boxplot(data=df_before[col], ax=axes[i][0], color='salmon')
    axes[i][0].set_title(f'{col} - Before')
    
    # After
    sns.boxplot(data=df[col], ax=axes[i][1], color='lightgreen')
    axes[i][1].set_title(f'{col} - After')

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
